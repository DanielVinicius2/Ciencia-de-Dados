```python
if 'Salario' in df_funcionarios.columns:
    salario = df_funcionarios['Salario'].dropna()

    Q1 = salario.quantile(0.25)
    Q2 = salario.quantile(0.50)
    Q3 = salario.quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = salario[(salario < limite_inferior) | (salario > limite_superior)]

    print(outliers)
else:
    print("Coluna não encontrada no dataset")
```
