# B. Tratamento de Dados Ausentes e Outliers

```python
# 1B
df_boardgame['avg_rating'].isnull().sum()
# Não há valores nulos
```

```python
# 2B
# Não há valores nulos.
#É mais recomendado preencher dados faltantes com a mediana do que com a média
#quando o conjunto de dados contém valores discrepantes (outliers) ou
#possui uma distribuição distorcida (assimétrica)
```

```python
# 3B
bplt_playingtime = px.box(df_boardgame, y='max_playtime')
bplt_playingtime.show()
# Acima do valor de 210 serão considerados outiliers
# Considerações:
# É possivel notar que a concentração de valores é maior entre Q1 e Q3 ou seja,
# O Intervalo Interquartil é de 45 a 120
# //////////////////////////////////////////////////////////////////////////////
# A mediana (60) encontrasse mais príxima do Q1 (45), logo é possivel notar que,
# a maior concentração de valores encontrasse acima da mediana(assimetria positiva).


```

```python
# 4B
Q1_maxplaytime = df_max_playtime.quantile(0.25)
Q3_maxplaytime = df_max_playtime.quantile(0.75)
IQR_maxplaytime = Q3_maxplaytime - Q1_maxplaytime
print(f"O Intervalo Interquartil é de {IQR_maxplaytime:.2f}")
```

# D. Análise da Relação entre Variáveis

```python
# 1D
scatter_plot_players = px.scatter(df_boardgame, x='min_players', y='max_players')
scatter_plot_players.show()
```

```python
# 2D
# A maioria dos jogos funciona para poucos jogadores (1 a 4 no mínimo),
# mas pode acomodar grupos maiores (até 10–20 no máximo).
# Existem alguns jogos atípicos que permitem até 100 jogadores, mas eles são exceções.
```

```python
# 3D
corr_matrix = df_boardgame.corr(numeric_only=True)
display(corr_matrix)
```

```python
# 4D
heat_map_corr_matrix = px.imshow(corr_matrix, text_auto=True)
heat_map_corr_matrix.show()
```

```python
# 5D
# Encontrar pares de correlações
corr_unstacked = corr_matrix.unstack().sort_values(ascending=False)

# Remover autocorrelações (coluna com ela mesma = 1.0)
corr_filtered = corr_unstacked[corr_unstacked < 1]

# Maior correlação positiva
max_corr = corr_filtered.head(1)

# Correlação mais próxima de zero
min_corr = corr_filtered.iloc[(corr_filtered - 0).abs().argsort()].head(1)

print("Maior correlação positiva:\n", max_corr)
print("\nCorrelação mais próxima de zero:\n", min_corr)
```

# E. Análise Temporal

```python
# 1E
df_boardgame_yearpublished = df_boardgame['release_year']
display(df_boardgame_yearpublished)
```

```python
# 2E
df_boardgame['Decada'] = (df_boardgame['release_year']// 10) * 10
print(df_boardgame[['release_year', 'Decada']].head())
```

```python
# 3E
jogos_por_decada = df_boardgame.groupby('Decada').size().reset_index(name='qtd_jogos')
print(jogos_por_decada)

df_decade_per_year = px.bar(jogos_por_decada,
             x='Decada',
             y='qtd_jogos')
df_decade_per_year.show()
```
